### testando conexão e definindo constantes

```python
ATENDIMENTO_TABLE = "`datario.adm_central_atendimento_1746.chamado`"
BAIRRO_TABLE = "`datario.dados_mestres.bairro`"
EVENTOS_TABLE = "`datario.turismo_fluxo_visitantes.rede_hoteleira_ocupacao_eventos`"
DATE = '2023-04-01'
PERTURBAÇAO = "Perturbação do sossego"
PERIODO_PERTURBAÇAO = "'2022-01-01' AND '2023-12-31'"

query = f"SELECT * FROM {ATENDIMENTO_TABLE} LIMIT 10"

def get_df(query):
  df = bd.read_sql(query, billing_project_id="desafio-rio")
  return display(df)

get_df(query)
```

#QUESTÃO 1 Quantos chamados foram abertos no dia 01/04/2023?



```python
query = f"""
SELECT COUNT(*) AS total_ocorrencias
FROM {ATENDIMENTO_TABLE}
WHERE DATE(data_inicio) = DATE('{DATE}')
"""

get_df(query)
```

#QUESTÃO 2 Qual o tipo de chamado que teve mais teve chamados abertos no dia 01/04/2023?

```python
query = f"""
SELECT tipo, COUNT(*) AS total_chamados
FROM
    {ATENDIMENTO_TABLE}
WHERE
    DATE(data_inicio) = DATE('{DATE}')
GROUP BY
    tipo
ORDER BY total_chamados DESC
LIMIT 1;
"""

get_df(query)
```

#QUESTÃO 3 Quais os nomes dos 3 bairros que mais tiveram chamados abertos nesse dia?

```python
query = f"""
SELECT b.nome AS nome_bairro, COUNT(c.id_bairro) AS total_chamados
FROM
    {ATENDIMENTO_TABLE} c
    JOIN {BAIRRO_TABLE} b ON c.id_bairro = b.id_bairro
WHERE
    DATE(c.data_inicio) = DATE('{DATE}')
GROUP BY
    b.nome
ORDER BY total_chamados DESC
LIMIT 3;
"""

get_df(query)
```

#QUESTÃO 4 Qual o nome da subprefeitura com mais chamados abertos nesse dia?

```python
query = f"""
SELECT b.subprefeitura, COUNT(c.id_bairro) AS total_chamados
FROM
    {ATENDIMENTO_TABLE} c
    JOIN {BAIRRO_TABLE} b ON c.id_bairro = b.id_bairro
WHERE
    DATE(c.data_inicio) = DATE('{DATE}')
GROUP BY
    b.subprefeitura
ORDER BY total_chamados DESC
LIMIT 1;
"""

get_df(query)
```

#QUESTÃO 5 Existe algum chamado aberto nesse dia que não foi associado a um bairro ou subprefeitura na tabela de bairros? Se sim, por que isso acontece?

```python
query = f"""
SELECT c.*
FROM
    `datario.adm_central_atendimento_1746.chamado` c
    LEFT JOIN `datario.dados_mestres.bairro` b ON c.id_bairro = b.id_bairro
WHERE
    DATE(c.data_inicio) = "2023-04-01"
    AND b.id_bairro IS NULL;
"""

get_df(query)
```

#QUESTÃO 6 Quantos chamados com o subtipo "Perturbação do sossego" foram abertos desde 01/01/2022 até 31/12/2023 (incluindo extremidades)?

```python
query = f"""
SELECT count(*) as chamados
FROM
    {ATENDIMENTO_TABLE}
WHERE
    subtipo = '{PERTURBAÇAO}'
    AND (
        DATE(data_inicio) BETWEEN {PERIODO_PERTURBAÇAO}
    );
"""

get_df(query)
```

#QUESTÃO 7 Selecione os chamados com esse subtipo que foram abertos durante os eventos contidos na tabela de eventos (Reveillon, Carnaval e Rock in Rio).

```python
query = f"""
SELECT COUNT(*) AS total_chamados
FROM
    {ATENDIMENTO_TABLE} c
    INNER JOIN {EVENTOS_TABLE} e ON (
        DATE(c.data_inicio) BETWEEN e.data_inicial AND e.data_final
    )
WHERE
    c.subtipo = "Perturbação do sossego";

"""

get_df(query)
```

#QUESTAO 8 Quantos chamados desse subtipo foram abertos em cada evento?

```python
query = f"""
SELECT e.evento, COUNT(c.id_chamado) AS total_chamados
FROM
    {ATENDIMENTO_TABLE} c
    INNER JOIN {EVENTOS_TABLE} e ON (
        DATE(c.data_inicio) BETWEEN e.data_inicial AND e.data_final
    )
WHERE
    c.subtipo = '{PERTURBAÇAO}'
GROUP BY
    e.evento
ORDER BY total_chamados DESC;
"""

get_df(query)
```

#QUESTAO 9 Qual evento teve a maior média diária de chamados abertos desse subtipo?

```python
subquery = f"""
SELECT
e.evento, DATE(c.data_inicio) AS data_inicio, DATE(c.data_fim) AS data_fim, COUNT(c.id_chamado) AS total_chamados
        FROM
            {ATENDIMENTO_TABLE} c
            INNER JOIN {EVENTOS_TABLE} e ON (
                DATE(c.data_inicio) BETWEEN e.data_inicial AND e.data_final
            )
        WHERE
            c.subtipo = "Perturbação do sossego"
        GROUP BY
            e.evento, data_inicio, data_fim
"""

query = f"""
SELECT subquery.evento, SUM(subquery.total_chamados) / (
        DATE_DIFF (
            MAX(subquery.data_fim), MIN(subquery.data_inicio), DAY
        ) + 1
    ) AS media_diaria
FROM (
        {subquery}
    ) AS subquery
GROUP BY
    subquery.evento
ORDER BY media_diaria DESC
LIMIT 1;
"""

get_df(query)
```

#QUESTAO 10 Compare as médias diárias de chamados abertos desse subtipo durante os eventos específicos (Reveillon, Carnaval e Rock in Rio) e a média diária de chamados abertos desse subtipo considerando todo o período de 01/01/2022 até 31/12/2023.

```python
subquery1 = f"""
media_eventos AS (
        SELECT subquery.evento, SUM(subquery.total_chamados) / (
                DATE_DIFF (
                    MAX(subquery.data_fim), MIN(subquery.data_inicio), DAY
                ) + 1
            ) AS media_diaria
        FROM (
                SELECT
                    e.evento, DATE(c.data_inicio) AS data_inicio, DATE(c.data_fim) AS data_fim, COUNT(c.id_chamado) AS total_chamados
                FROM
                    {ATENDIMENTO_TABLE} c
                    INNER JOIN {EVENTOS_TABLE} e ON (
                        DATE(c.data_inicio) BETWEEN e.data_inicial AND e.data_final
                    )
                WHERE
                    c.subtipo = '{PERTURBAÇAO}'
                GROUP BY
                    e.evento, data_inicio, data_fim
            ) AS subquery
        GROUP BY
            subquery.evento
    )
"""

subquery2 = f"""
media_total AS (
        SELECT COUNT(*) / (
                DATE_DIFF (
                    DATE '2023-12-31', DATE '2022-01-01', DAY
                ) + 1
            ) AS media_diaria
        FROM datario.adm_central_atendimento_1746.chamado
        WHERE
            subtipo = 'Perturbação do sossego'
            AND DATE(data_inicio) BETWEEN {PERIODO_PERTURBAÇAO}
    )
"""

query = f"""
WITH
    {subquery1}
    ,
    {subquery2}
SELECT
    media_eventos.evento AS evento,
    media_eventos.media_diaria AS media_diaria_evento,
    media_total.media_diaria AS media_diaria_total
FROM media_eventos, media_total;
"""

get_df(query)
```
